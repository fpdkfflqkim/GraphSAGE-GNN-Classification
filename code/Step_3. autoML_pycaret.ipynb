```python
# Import necessary libraries
import pandas as pd  

# PyCaret for AutoML modeling
from pycaret.classification import *  # Import all necessary modules for classification tasks
```

# 데이터 불러오기 및 전처리

```python
train = pd.read_csv('./data/0903_full_emb/full_0903_20학습_20추론_emb.csv').drop(columns='br_num_year')
test = pd.read_csv('./data/0903_full_emb/full_0903_20학습_21추론_emb.csv').drop(columns='br_num_year')
X_train = train.drop(columns=['label','label_1'])

X_test = test.drop(columns=['label','label_1'])
```

```python
# Display the distribution of predictions in the test data
X_train['label_2'].value_counts()/len(train)
```

```python
scaler = RobustScaler()
X_train.iloc[:,:-1] = scaler.fit_transform(X_train.iloc[:,:-1])
X_test.iloc[:,:-1] = scaler.transform(X_test.iloc[:,:-1])
```

# 모델 비교

```python
# Initialize the PyCaret environment and define the setup
s = setup(X_train, target = 'label_2', session_id=42)

best = compare_models()
```

# 최적 모델 생성

```python
# Create and tune a model using PyCaret's AutoML
best_model  = create_model('rf')

tuned_model = tune_model(best_model)

predicted = predict_model(tuned_model, data=X_test)
predicted
```

```python
# Display the distribution of predictions in the test data
predicted.prediction_label.value_counts()/len(predicted)
```
