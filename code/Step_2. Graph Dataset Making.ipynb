```python
import pandas as pd

ab = pd.read_csv('./data/apollo/240904_전체년도ab_결측치이상치처리완료(3년후매출액증가율)_v4.csv',encoding='utf-8')

bg = pd.read_csv('./data/apollo/bg03.csv')
```

```python
bg['br_num'] = bg['br_num'].astype('Int64')
bg['cp_br_num'] = bg['cp_br_num'].astype('Int64')
```

```python
bg.dropna(subset='sup_amt', inplace=True)
bg.dropna(subset='br_num', inplace=True)
bg.dropna(subset='cp_br_num', inplace=True)
```

```python
bg_origin= bg[bg['data_typ'].isin([1,3,5,7])]
bg_reverse = bg[bg['data_typ'].isin([2,4,6,8])]
bg_reverse.rename(columns={'br_num':'cp_br_num2'}, inplace=True)
bg_reverse.rename(columns={'cp_br_num':'br_num'}, inplace=True)
bg_reverse.rename(columns={'cp_br_num2':'cp_br_num'}, inplace=True)
bg_origin = bg_origin[['br_num','txn_end','cp_br_num','sup_amt']]
bg_reverse = bg_reverse[['br_num','txn_end','cp_br_num','sup_amt']]
bg_all = pd.concat([bg_origin,bg_reverse])
bg_all
```

```python
# bg_all['br_num'] = bg_all['br_num'].astype('int16')
bg_all['br_num'] = bg_all['br_num'].astype(str)
bg_all['txn_end'] = bg_all['txn_end'].astype(str)
bg_all['br_num_year'] = bg_all['br_num'].astype(str) + "_" + bg_all['txn_end'].str.slice(start=2, stop=4)
# bg_all['cp_br_num'] = bg_all['cp_br_num'].astype('int16')
bg_all['cp_br_num'] = bg_all['cp_br_num'].astype(str)
bg_all['cp_br_num_year'] = bg_all['cp_br_num'].astype(str) + "_" + bg_all['txn_end'].str.slice(start=2, stop=4)
bg_all
```

```python
bg_all = bg_all.groupby(['br_num_year','cp_br_num_year'], as_index=False)['sup_amt'].sum()
```

```python
#  Edge attr : 거래액 / 매출액

ab_all_mce = ab_all[['br_num_year','매출액']]
bg_all = pd.merge(bg_all,ab_all_mce, how='left', on='br_num_year')
bg_all['edge_attr'] = bg_all['sup_amt'] / bg_all['매출액']
bg_all.drop(columns=['sup_amt','매출액'], inplace=True)
```

```python
#### ab, bg 매핑
bg = bg_all
# ab에 없는 기업 bg에서 날리기
ab['br_num_year'] = ab['br_num_year'].astype(str)
bg['br_num_year'] = bg['br_num_year'].astype(str)
bg['cp_br_num_year'] = bg['cp_br_num_year'].astype(str)
abli = list(ab['br_num_year'].unique())
bg = bg[(bg['br_num_year'].isin(abli))&(bg['cp_br_num_year'].isin(abli))]

# bg에 없는 기업 ab에서 날리기
bgli = list(set(list(bg['br_num_year'])+list(bg['cp_br_num_year'])))
# 거래 존재 기업 노드 ab_gr
ab_gr = ab[ab['br_num_year'].isin(bgli)]
```

```python
# 풀 데이터


ab_year = ab_gr[ab_gr['year']=='18']
ab_ngr_year = ab_ngr[ab_ngr['year']=='18']
ab_ngr_year.drop(columns='year',inplace=True)

bg_year = bg[bg['year']=='18']

unique_companies = pd.unique(bg_year[['br_num_year', 'cp_br_num_year']].values.ravel('K'))
company_to_index = {company: idx for idx, company in enumerate(unique_companies)}
bg_year['put'] = bg_year['br_num_year'].map(company_to_index)
bg_year['get'] = bg_year['cp_br_num_year'].map(company_to_index)
bg_year = bg_year.sort_values(by=['put', 'get']).reset_index(drop=True)
# 재무 데이터 매칭
ab_year['idx'] = ab_year['br_num_year'].map(company_to_index)
ab_year.sort_values(by='idx', inplace=True)

# 노드(피처, 라벨) / 엣지(인덱스, 어트리뷰트) 저장
# 노드 저장
ab_year = ab_year.iloc[:, :-1]

# 엣지 저장
bg_year = bg_year[['put','get','edge_attr']]

bg_year['put'] = bg_year['put'].astype('Int64')
bg_year['get'] = bg_year['get'].astype('Int64')
bg_year['edge_attr'] = bg_year['edge_attr'].astype('float64')
print(f'18년도 거래 수 : {len(bg_year)}, 기업 수 : {len(ab_year)}')
bg_yearli_20_all = list(set(list(bg_year['put'].unique()) + list(bg_year['get'].unique())))
bg_year = bg_year.transpose()

if len(ab_year) != len(bg_yearli_20_all) :
    print(f'비상!!! 노드 수 : {len(ab_year)}, 엣지 기업 수 : {len(bg_yearli_20_all)}')

ab_all = pd.concat([ab_year,ab_ngr_year],ignore_index=True)
print(len(ab_all))
ab_all
# ab_all.to_csv(f'./data/apollo/Inductive dataset/240910_{year}_full_node(3년후).csv', encoding='utf-8', index=False)
# bg_year.to_csv(f'./data/apollo/Inductive dataset/240910_{year}_full_edge(3년후).csv', encoding='utf-8', index=False)
```

```python
# 풀 데이터

yearli = ['18','19','20','21']

for year in yearli:

    ab_year = ab_gr[ab_gr['year']==year]
    ab_ngr_year = ab_ngr[ab_ngr['year']==year]
    ab_ngr_year.drop(columns='year',inplace=True)
    ab_year.drop(columns='year',inplace=True)
    bg_year = bg[bg['year']==year]

    unique_companies = pd.unique(bg_year[['br_num_year', 'cp_br_num_year']].values.ravel('K'))
    company_to_index = {company: idx for idx, company in enumerate(unique_companies)}
    bg_year['put'] = bg_year['br_num_year'].map(company_to_index)
    bg_year['get'] = bg_year['cp_br_num_year'].map(company_to_index)
    bg_year = bg_year.sort_values(by=['put', 'get']).reset_index(drop=True)
    # 재무 데이터 매칭
    ab_year['idx'] = ab_year['br_num_year'].map(company_to_index)
    ab_year.sort_values(by='idx', inplace=True)

    # 노드(피처, 라벨) / 엣지(인덱스, 어트리뷰트) 저장
    # 노드 저장
    ab_year = ab_year.iloc[:, :-1]

    # 엣지 저장
    bg_year = bg_year[['put','get','edge_attr']]

    bg_year['put'] = bg_year['put'].astype('Int64')
    bg_year['get'] = bg_year['get'].astype('Int64')
    bg_year['edge_attr'] = bg_year['edge_attr'].astype('float64')
    print(f'{year}년도 거래 수 : {len(bg_year)}, 기업 수 : {len(ab_year)}')
    bg_yearli_20_all = list(set(list(bg_year['put'].unique()) + list(bg_year['get'].unique())))
    bg_year = bg_year.transpose()
    
    if len(ab_year) != len(bg_yearli_20_all) :
        print(f'비상!!! 노드 수 : {len(ab_year)}, 엣지 기업 수 : {len(bg_yearli_20_all)}')
    
    ab_all = pd.concat([ab_year,ab_ngr_year],ignore_index=True)
    print(len(ab_all))
    ab_all.to_csv(f'./data/apollo/Inductive dataset/240910_{year}_full_node(3년후).csv', encoding='utf-8', index=False)
    bg_year.to_csv(f'./data/apollo/Inductive dataset/240910_{year}_full_edge(3년후).csv', encoding='utf-8', index=False)
```
